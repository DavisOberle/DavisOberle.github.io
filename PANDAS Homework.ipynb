```python
import pandas as pd
import numpy as np

file_to_load = "Desktop/purchase_data.csv"

purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
total_players = len(purchase_data["SN"].value_counts())
total_players

pd.DataFrame({"Total Players": [total_players]})
```

```python
unique_items = len(purchase_data["Item ID"].value_counts())
unique_items

average = purchase_data["Price"].mean()
average               

individual_purchases = purchase_data["Price"].count()
individual_purchases

purchases_total = purchase_data["Price"].sum()
purchases_total

pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [average], 
              "Number of Purchases": individual_purchases, "Total": [purchases_total]})
```

```python
total_male = purchase_data.Gender == "Male"
male = total_male.sum()
total_female = purchase_data.Gender == "Female"
female = total_female.sum()
total_other = purchase_data.Gender == "Other / Non-Disclosed"
other = total_other.sum()


pd.DataFrame({"Total Count": [male, female, other], 
              "Percentage of Players": [male/780, female/780, other/780]}, 
             index = ["Male", "Female", "Other"])
```

```python
female_avg = purchase_data.loc[purchase_data["Gender"] == "Female"]
female_avg2 = female_avg["Price"].mean()
female_total = female_avg["Price"].sum()


male_avg = purchase_data.loc[purchase_data["Gender"] == "Male"]
male_avg2 = male_avg["Price"].mean()
male_total = male_avg["Price"].sum()

other_avg = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed"]
other_avg2 = other_avg["Price"].mean()
other_total = other_avg["Price"].sum()
```

```python
gender_df = pd.DataFrame({"Purchase Count": [male, female, other],
                         "Average Purcahse Price": [male_avg2, female_avg2, other_avg2],
                         "Total Purchase Value": [male_total, female_total, other_total],
                         "Avg Total Purchase per Person": [male/male_total, female/female_total, other/other_total]})

gender_df
```

```python
bins = [0, 10, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
purchase_data["Age Summary"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

purchase_data["Age Summary"]

age_count = purchase_data["Age Summary"].value_counts()

age_group = purchase_data.groupby("Age Summary")

age_percentage = age_count/780

age_df = pd.DataFrame({"Total Count": age_count, 
                      "Percentage of Players": age_percentage},
                     index = group_names)
age_df
```

```python
average_purchase_price = age_group.Price.mean()
total_age_price = age_group.Price.sum()
avg_total_per_person = age_count/total_age_price

age_analysis = pd.DataFrame({"Purchase Count": age_count, 
                            "Average Purchase Price": average_purchase_price,
                            "Total Purchase Value": total_age_price,
                            "Avg Total Purchase per Person": avg_total_per_person},
                             index = group_names)
age_analysis


```
